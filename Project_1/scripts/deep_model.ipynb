```python
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense, Conv2D
from datetime import date, timedelta, datetime
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv ('Merged.csv')
```

```python
my_list = list(df)

print(len(my_list))
```

```python
#zmiana kolumny i sprawdzenie czy dobrze się wgrały
nazwy_kolumn = ['data', 'dzien_roboczy','swieto','ferie'
             ,'wilgotonosc_East','wilgotonosc_North','wilgotonosc_South','wilgotonosc_Warsasw','wilgotonosc_Wast'
             ,'predkosc_wiatru_East','predkosc_wiatru_North','predkosc_wiatru_South','predkosc_wiatru_Warsasw','predkosc_wiatru_Wast' 
             ,'max_temp_East','max_temp_North','max_temp_South','max_temp_Warsasw','max_temp_Wast'
             ,'opady_East','opady_North','opady_South','opady_Warsasw','opady_Wast'
             ,'temp_East','temp_North','temp_South','temp_Warsasw','temp_Wast'
             ,'wilgotnosc_East','wilgotnosc_North','wilgotnosc_South','wilgotnosc_Warsasw','wilgotnosc_Wast'
             ,'wiatr_East','wiatr_North','wiatr_South','wiatr_Warsasw','wiatr_Wast'
             ,'zachmurz_East','zachmurz_North','zachmurz_South','zachmurz_Warsasw','zachmurz_Wast'
             ,'widzialnosc_East','widzialnosc_North','widzialnosc_South','widzialnosc_Warsasw','widzialnosc_Wast'
             ,'pm10_Eest','pm10_North','pm10_South','pm10_Warsaw','pm10_West'
            ,'nox_Eest','nox_North','nox_South','nox_Warsaw','nox_West' 
             ,'co_East','co_North','co_South','co_Warsaw','co_West'
             ,'pm2_5_Eest','pm2_5_North','pm2_5_South','pm2_5_Warsaw','pm2_5_West'] 


for i,name in enumerate(my_list):
    print(nazwy_kolumn[i],'<------->',name)
```

```python
#zmiana nazw kolumn
df.columns = nazwy_kolumn
cos = list(df['pm2_5_Warsaw'])
df
```

```python
# dodanie miesiaca
df['month'] = pd.DatetimeIndex(df['data']).month
df
```

```python
# wybor danych z odpowiendich miesięcy
list_of_months = [10,11,12,1,2]

#wybór wierszy z 2019 lub niższego roku
sdate = date(2017, 1, 1)   # start date
edate = date(2019, 12, 31)   # end date
delta = edate - sdate       # as timedelta

list_of_days =[]

for i in range(delta.days + 1):
    day = sdate + timedelta(days=i)
    day = str(day)
    list_of_days.append(day)




df_half = df[df.month.isin(list_of_months)]
df_half = df[df.data.isin(list_of_days)]
df_half
```

```python
# Model 1: pogoda jutro -> zanieczyszczenie jutro
```

```python
#dane tylko dla warszawy

list_of_values = ['dzien_roboczy','swieto','ferie',
                  'wilgotonosc_Warsasw', 'predkosc_wiatru_Warsasw', 'max_temp_Warsasw', 'opady_Warsasw','wilgotnosc_Warsasw', 'wiatr_Warsasw', 'zachmurz_Warsasw','widzialnosc_Warsasw'
                    ,'pm10_Warsaw']
n_shape = len(list_of_values) -1

df_warsaw = df_half[list_of_values]


#usuniecie danych z pustymi wartosciami
df_warsaw = df_warsaw.dropna()
df_warsaw = df_warsaw.astype(int)


#stworzenie zbiorów do uczenia
X_data = df_warsaw[list_of_values[:-1]]
y_data = df_warsaw['pm10_Warsaw']


y_data = pd.DataFrame(y_data).astype(int)
```

```python
#zmiana Y na 0/1 w zalezności od porgu odcięcia

prog_odciecia = 30
y_data_with_numb = y_data

y_data.loc[y_data['pm10_Warsaw'] >= prog_odciecia, 'czy_smog_0_1'] = 1
y_data.loc[y_data['pm10_Warsaw'] < prog_odciecia, 'czy_smog_0_1'] = 0
y_data = y_data['czy_smog_0_1']
y_data = y_data.values
y_data
```

```python
# przygotowanie danych i preprocesing


X_data = X_data.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
X_scaled = min_max_scaler.fit_transform(X_data)
X_scaled
```

```python
#podzial na test i train 

x_train, x_test, y_train, y_test = train_test_split(X_scaled, y_data)
```

```python
model = Sequential()


model.add(Dense(40, input_dim=n_shape, activation='relu'))
model.add(Dense(20, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

model.fit(x_train, y_train, epochs=500, batch_size=10)
```

```python
accuracy = model.evaluate(x_test, y_test)
accuracy[1]
```

```python
# Model 2: pogoda i zanieczyszczenie dziś i n dni wstecz -> zanieczyszczenie jutro
```

```python
#dane tylko dla warszawy

list_of_values = ['data','month',
                    'dzien_roboczy','swieto','ferie',
                  'wilgotonosc_Warsasw', 'predkosc_wiatru_Warsasw', 'max_temp_Warsasw', 'opady_Warsasw','wilgotnosc_Warsasw', 'wiatr_Warsasw', 'zachmurz_Warsasw','widzialnosc_Warsasw'
                    ,'co_Warsaw','nox_Warsaw','pm10_Warsaw']

df_warsaw = df_half[list_of_values]
df_warsaw  = df_warsaw.dropna()

df_all_warsaw = df[list_of_values]
df_all_warsaw = df_all_warsaw.dropna()

df_all_warsaw
# df
```

```python
target = df_all_warsaw['pm10_Warsaw']
target


#dodanie laga n poprzednich dni 
max_lag = 3
datas = []
for l in range(1, max_lag+1):
  lagged_data = df_all_warsaw.shift(l)
  lagged_data = lagged_data.rename({c: f"{c}_L{l}" for c in df_all_warsaw.columns}, axis=1)
  datas.append(lagged_data)
df_all_warsaw = pd.concat(datas, axis=1)
df_all_warsaw['target jutro'] = target.shift(-1)


df_all_warsaw = df_all_warsaw.dropna()
df_all_warsaw
```

```python
nazwy_kolumn = list(df_all_warsaw.columns)
# nazwy_kolumn

for i in range(1, max_lag+1):
    nazwy_kolumn.remove(f"data_L{i}")
    nazwy_kolumn.remove(f"month_L{i}")
    print(f"data_L{i}")
    
#stworzenie zbiorów do uczenia
X_data = df_all_warsaw[nazwy_kolumn[:-1]]
y_data = df_all_warsaw['target jutro']


y_data = pd.DataFrame(y_data).astype(int)
y_data
```

```python
# przygotowanie danych i preprocesing


X_data = X_data.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
X_scaled = min_max_scaler.fit_transform(X_data)
X_scaled
```

```python
prog_odciecia = 30

y_data.loc[y_data['target jutro'] >= prog_odciecia, 'czy_smog_0_1'] = 1
y_data.loc[y_data['target jutro'] < prog_odciecia, 'czy_smog_0_1'] = 0
y_data = y_data['czy_smog_0_1']
y_data = y_data.values
y_data
```

```python
#podzial na test i train 

x_train, x_test, y_train, y_test = train_test_split(X_scaled, y_data)
```

```python
model_lag = Sequential()

n_shape = x_train.shape[1]

model_lag.add(Dense(40, input_dim=n_shape, activation='relu'))
model_lag.add(Dense(20, activation='relu'))
model_lag.add(Dense(1, activation='sigmoid'))

model_lag.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

model_lag.fit(x_train, y_train, epochs=50, batch_size=10)
```

```python
accuracy = model_lag.evaluate(x_test, y_test)
accuracy[1]
```
